#### Begin Input Read

```python
#root directory
rt_path = f'//hecate/Insurance_US/Product Development/Product Management/Global PPM/Reporting/Data Collection/{str(percode)}/Manual Updates/Swiss Files'
#json input directory

pq_input = os.path.join( rt_path, 'us_dat_CH_06042021132400.parquet')
map_path = os.path.join( rt_path, 'ch_dc_bp.csv')
pq_output = os.path.join( rt_path,'Output' ,'us_dat_CH_06042021132401.parquet')

pq_input
```

```python
#Display Options
pd.set_option('display.max_rows', 500)
```

```python
ch_df =  pd.read_parquet(pq_input, engine="pyarrow")
map_df = pd.read_csv(map_path,  encoding='utf-8', engine='python')
```

```python
map_df
```

```python
print(len(ch_df))

fixed_dc = ch_df.merge(map_df, how = 'left',left_on='Business Partner Name', right_on='business_partner_name')

fixed_dc['Distribution Channel'] = fixed_dc['distribution_channel'].fillna("Not Available")

pd.set_option('display.max_rows', 5000)
fixed_dc['Distribution Channel']
len(fixed_dc)

fixed_dc = fixed_dc.drop(['country', 'distribution_channel', 'business_partner_name'], axis=1)

fixed_dc.columns

```

```python
fixed_dc.to_parquet(pq_output, engine="pyarrow")
```
